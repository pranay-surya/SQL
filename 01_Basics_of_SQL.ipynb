## SQL Fundamentals 

### What is Database ?
    A Database is an organized collection of structured information, or data, typically stored electronically in a computer system. Think of it as a digital filing cabinet. While a spreadsheet (like Excel) is great for one person to calculate numbers, a database is designed to hold massive amounts of data that many users can access and update at the same time securely.

### What is SQL
    SQL (pronounced "sequel" or "S-Q-L") is the standard programming language used to communicate with and manage a relational database. If the database is the filing cabinet, SQL is the language you use to tell the cabinet to "give me all files from 2023" or "add a new file for customer John Doe."

### Create table students 

```python
%%sql 

CREATE TABLE students (
    student_id INT PRIMARY KEY,
    name VARCHAR(50),
    age INT,
    gender CHAR(1),
    course VARCHAR(50),
    score INT,
    join_date DATE
);
```
